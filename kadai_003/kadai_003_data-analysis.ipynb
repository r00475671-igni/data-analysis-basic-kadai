```python
import numpy as np

multi_arrayA = np.array([[0,1],[2,3],[4,5]])

import numpy as np

multi_arrayA = np.array([[0,1],[2,3],[4,5]])

multi_arrayB = np.array([[0,1,2,3],[4,5,6,7]])

print(f'"A=" {multi_arrayA}')

print(f'"B=" {multi_arrayB}')
```

```python
#np.dot(a,b)で行列aと行列bの積

result = np.dot(multi_arrayA, multi_arrayB)

print(result)
```

```python
#計算された行列の最大値を求めて表示

result.max()
```
